```python
# 1. წაკითხვა
test_df = pd.read_csv('/kaggle/input/datasets/tamarigelashvili/house-pricess/house-prices-advanced-regression-techniques/test.csv')
test_ids = test_df['Id']

# 2. ჩვენი შექმნილი Feature-ების დამატება
test_df['HouseAge'] = test_df['YrSold'] - test_df['YearBuilt']

# 3. კატეგორიული ცვლადების გადაყვანა (Mapping), რაც ექსპერიმენტში ვქენით
qual_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0}
qual_cols = ['ExterQual', 'BsmtQual', 'HeatingQC', 'KitchenQual', 'FireplaceQu']

for col in qual_cols:
    test_df[col] = test_df[col].map(qual_map).fillna(0)

# 4. One-Hot Encoding-ის სვეტების ხელით შექმნა (რომელიც Lasso-მ დატოვა)
test_df['MSZoning_RM'] = (test_df['MSZoning'] == 'RM').astype(int)
test_df['LotConfig_CulDSac'] = (test_df['LotConfig'] == 'CulDSac').astype(int)
test_df['Neighborhood_NoRidge'] = (test_df['Neighborhood'] == 'NoRidge').astype(int)
test_df['Condition1_Norm'] = (test_df['Condition1'] == 'Norm').astype(int)
test_df['RoofStyle_Gable'] = (test_df['RoofStyle'] == 'Gable').astype(int)
test_df['MasVnrType_Stone'] = (test_df['MasVnrType'] == 'Stone').astype(int)
test_df['BsmtExposure_Gd'] = (test_df['BsmtExposure'] == 'Gd').astype(int)
test_df['BsmtExposure_No'] = (test_df['BsmtExposure'] == 'No').astype(int)
test_df['BsmtFinType1_GLQ'] = (test_df['BsmtFinType1'] == 'GLQ').astype(int)
test_df['BsmtFinType1_Unf'] = (test_df['BsmtFinType1'] == 'Unf').astype(int)
test_df['GarageType_Detchd'] = (test_df['GarageType'] == 'Detchd').astype(int)
test_df['SaleType_New'] = (test_df['SaleType'] == 'New').astype(int)

# 5. მოდელის მიერ მოთხოვნილი სვეტების ზუსტი სია და თანმიმდევრობა
expected_features = [
    'MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 
    'MasVnrArea', 'ExterQual', 'BsmtQual', 'HeatingQC', '1stFlrSF', 
    'GrLivArea', 'BsmtFullBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 
    'KitchenQual', 'TotRmsAbvGrd', 'Fireplaces', 'FireplaceQu', 'GarageCars', 
    'WoodDeckSF', 'HouseAge', 'MSZoning_RM', 'LotConfig_CulDSac', 
    'Neighborhood_NoRidge', 'Condition1_Norm', 'RoofStyle_Gable', 
    'MasVnrType_Stone', 'BsmtExposure_Gd', 'BsmtExposure_No', 
    'BsmtFinType1_GLQ', 'BsmtFinType1_Unf', 'GarageType_Detchd', 'SaleType_New'
]

# 6. საბოლოო ცხრილის მომზადება
X_test_kaggle = test_df[expected_features].copy()
X_test_kaggle = X_test_kaggle.fillna(X_test_kaggle.median())

# 7. პროგნოზი
predictions = loaded_model.predict(X_test_kaggle)
```

```python
submission = pd.DataFrame({
    'Id': test_ids,
    'SalePrice': predictions
})

submission.to_csv('submission_xgboost.csv', index=False)
print("✅ ფაილი 'submission_xgboost.csv' წარმატებით შეიქმნა!")
```
